```python
from bs4 import BeautifulSoup
import requests
url= BeautifulSoup('https://www.worldometers.info/world-population/population-by-country/', 'html.parser')
soup = requests.get(url)
soup
```

```python
src = soup.content
document = BeautifulSoup(src, "lxml")
soup
```

```python
table = document.find("table")
table
```

```python
rows = table.find_all("tr")
rows
```

```python
for row in rows[1:-1]:
    cells = row.find_all(["td", "th"])
    
    cells_text = [cell.get_text() for cell in cells]
    print(cells_text)
```
